```python
from music21 import *
import glob
notes = []
from roll import MidiFile
```

```python
musicFile = converter.parse ('media/test_output/test_output.mid')
```

```python
musicFile.show ()
```

```python
try :
    inst = instrument.partitionByInstrument (musicFile)
    notesToParse = inst.parts[0].recurse ()
except :
    notesToParse = musicFile.flat.notes
for ele in notesToParse :
    print (ele)
    if isinstance (ele, note.Note) :
        notes.append (str (ele.pitch))
    elif isinstance (ele, chord.Chord) :
        notes.append ('.'.join(str(n) for n in ele.normalOrder))
```

```python
sorted (set (notes))
```

```python
noteCoding = {}
pitchNames = set (notes)
nNotes = len (pitchNames)
pitchNames = sorted (pitchNames)
code = 0
for i in pitchNames :
    noteCoding [i] = code
    code += 1
```

```python
import matplotlib.pyplot as plt
import librosa.display
import librosa
import numpy as np
y, sr = librosa.load('media/test_output/test_output.wav')
```

```python
C = librosa.feature.chroma_cqt(y=y, sr=sr)
#plt.subplot(4, 2, 5)
librosa.display.specshow(C, y_axis='chroma')
plt.colorbar()
plt.title('Chromagram')
```

```python
CQT = librosa.amplitude_to_db(np.abs(librosa.cqt(y, sr=sr)), ref=np.max)
librosa.display.specshow(CQT, y_axis='cqt_note')
plt.colorbar(format='%+2.0f dB')
plt.title('Constant-Q power spectrogram (note)')
```
